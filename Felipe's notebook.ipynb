```python
import pandas as pd
education = pd.read_csv("Education.csv")
education
```

```python
enrollment = dict()

for index, row in education.iterrows():
    place = row['neighborhood']
    students = row['total_students_enrolled']
    if students in enrollment:
        enrollment[place] = students
    else:
        enrollment[place] = students
```

##### I created this dictionary in order to drop all of the columns that I did not need for my data set.

```python
newenrollment = pd.DataFrame.from_dict(enrollment, orient = 'index', columns = ['students'])
newenrollment
```
